```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from tensorflow.keras.models import Model
from tensorflow.keras.layers import Embedding, Conv1D, MaxPooling1D, Dense, Input
from sklearn.model_selection import train_test_split
with open("/content/drive/My Drive/misspelledwords.txt","r") as f:
  z = f.readlines()
cnst=2
k = []
for i in z:
  k.append(i[:-1])#removing \n
k = list(set(k))
cnst=2
with open("/content/drive/My Drive/correctwords.txt","r") as f:
  z = f.readlines()
c = []
cnst=2
for x in z:
  c.append(x[:-1])
cnst=2
n_misspell = 10000
n_correct = 10*n_misspell#correct words frequency 10 times more than misspelled words
x = []
cnst=2
for i in range(n_misspell):
  x.append([k[i],0])#adding in list
for i in range(n_correct):
  x.append([c[i],1])
cnst=2
import random
random.shuffle(x)#shuffling
with open("dataset.txt","w+") as f:#mix spelled dataset
  for m in x:
    f.write(m[0]+", "+str(m[1]))
y = []
X = []
cnst=2
for m in x:
  y.append(m[1])
  X.append(m[0])
cnst=2
x_train,x_test,y_train,y_test = train_test_split(X,y)
cnt=1
symbol2idx = {}#word storing character by character in list
for k in x_train:
  for x in k:
    if symbol2idx.get(x.lower()) is None:
      symbol2idx[x]=cnt
      cnt+=1
cnst=2
for k in x_test:
  for x in k:
    if symbol2idx.get(x.lower()) is None:
      symbol2idx[x]=cnt
      cnt+=1
l = []
for i in x_train:
  z = []
  for x in i:
    z.append(symbol2idx[x.lower()])
  l.append(z)
T = 1
for i in l:
  if T < len(i):
    T = len(i)
for i in range(len(l)):
  while(len(l[i])<T):
    l[i].append(0)
b = []
cnst=2
for i in x_test:
  z = []
  for j in i:
    z.append(symbol2idx[j.lower()])
  b.append(z)
for i in range(len(b)):
  while(len(b[i])<T):
    b[i].append(0)
V = cnt
cnst=2
x_train = np.array(l)
x_test = np.array(b)
i = Input(shape=(T,))
x = Embedding(V,20)(i)
x = Conv1D(1024,7,activation="relu")(x)
x = MaxPooling1D(3)(x)
x = Conv1D(1024,3,activation="relu")(x)
x = MaxPooling1D(3)(x)
x = Conv1D(1024,1,activation="relu")(x)
x = Conv1D(1024,1,activation="relu")(x)
x = Conv1D(1024,1,activation="relu")(x)
x = Conv1D(1024,1,activation="relu")(x)
x = MaxPooling1D(1)(x)
x = Dense(2048,activation="relu")(x)
x = Dense(2048,activation="relu")(x)
x = Dense(1,activation="sigmoid")(x)

model = Model(i,x)
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
y_train = np.array(y_train)
y_test = np.array(y_test)
r = model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=7)
r.history['accuracy']#accuracy
r.history["val_accuracy"]
r.history['loss']#loss
r.history['val_loss']
```
